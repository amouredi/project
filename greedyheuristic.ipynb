```python
import os
import itertools
import pandas as pd

# ============================================================
# 1) INPUT DATA
# ============================================================

# -----------------------------
# Team base-camp cities
# -----------------------------
team_base_city = {
    "Morocco": "Rabat",
    "Zambia": "Casablanca",
    "Mali": "Casablanca",
    "Comoros": "Casablanca",
    "Angola": "Marrakesh",
    "South Africa": "Marrakesh",
    "Egypt": "Agadir",
    "Zimbabwe": "Marrakesh",
    "Nigeria": "Fez",
    "Tunisia": "Rabat",
    "Uganda": "Rabat",
    "Tanzania": "Rabat",
    "Senegal": "Tangier",
    "Congo": "Rabat",
    "DR Congo": "Rabat",
    "Benin": "Rabat",
    "Botswana": "Rabat",
    "Algeria": "Rabat",
    "Burkina Faso": "Casablanca",
    "Equatorial Guinea": "Casablanca",
    "Sudan": "Casablanca",
    "Cote d'Ivoire": "Marrakesh",
    "Cameroon": "Agadir",
    "Gabon": "Agadir",
    "Mozambique": "Agadir",
}

# -----------------------------
# Stadiums
# -----------------------------
stadiums_df = pd.DataFrame([
    {"stadium": "Grand Stadium of Agadir", "city": "Agadir", "capacity": 45480},
    {"stadium": "Mohammed V Complex", "city": "Casablanca", "capacity": 45000},
    {"stadium": "Grand Stadium of Fez", "city": "Fez", "capacity": 45000},
    {"stadium": "Grand Stadium of Marrakesh", "city": "Marrakesh", "capacity": 45240},
    {"stadium": "Grand Stadium of Tangier", "city": "Tangier", "capacity": 75600},
    {"stadium": "Prince Moulay Abdellah Sports Complex", "city": "Rabat", "capacity": 69500},
    {"stadium": "Moulay El Hassan II Stadium", "city": "Rabat", "capacity": 22000},
    {"stadium": "Stade Al Medina", "city": "Rabat", "capacity": 21000},
    {"stadium": "Rabat Olympic Stadium", "city": "Rabat", "capacity": 21000},
])

# -----------------------------
# One-way travel time matrix (hours)
# -----------------------------
travel_time = {
    "Rabat": {
        "Rabat": 0.00, "Casablanca": 1.37, "Agadir": 5.57,
        "Marrakesh": 3.50, "Fez": 2.40, "Tangier": 2.83
    },
    "Casablanca": {
        "Rabat": 1.37, "Casablanca": 0.00, "Agadir": 4.80,
        "Marrakesh": 2.47, "Fez": 3.03, "Tangier": 3.77
    },
    "Agadir": {
        "Rabat": 5.57, "Casablanca": 4.80, "Agadir": 0.00,
        "Marrakesh": 2.62, "Fez": 7.27, "Tangier": 7.87
    },
    "Marrakesh": {
        "Rabat": 3.50, "Casablanca": 2.47, "Agadir": 2.62,
        "Marrakesh": 0.00, "Fez": 5.50, "Tangier": 5.90
    },
    "Fez": {
        "Rabat": 2.40, "Casablanca": 3.03, "Agadir": 7.27,
        "Marrakesh": 5.50, "Fez": 0.00, "Tangier": 4.37
    },
    "Tangier": {
        "Rabat": 2.83, "Casablanca": 3.77, "Agadir": 7.87,
        "Marrakesh": 5.90, "Fez": 4.37, "Tangier": 0.00
    },
}

# ============================================================
# 2) MATCH TABLES
# ============================================================

groups = {
    "A": ["Morocco", "Zambia", "Mali", "Comoros"],
    "B": ["Angola", "South Africa", "Egypt", "Zimbabwe"],
    "C": ["Nigeria", "Tunisia", "Uganda", "Tanzania"],
    "D": ["Senegal", "Congo", "Benin", "Botswana"],
    "E": ["Algeria", "Burkina Faso", "Equatorial Guinea", "Sudan"],
    "F": ["Cote d'Ivoire", "Cameroon", "Gabon", "Mozambique"],
}

# Group stage
group_matches = []
for group_name, team_list in groups.items():
    for k, (t1, t2) in enumerate(itertools.combinations(team_list, 2), start=1):
        group_matches.append({
            "match_id": f"{group_name}{k}",
            "round": "Group Stage",
            "team1": t1,
            "team2": t2
        })

group_matches_df = pd.DataFrame(group_matches)

# Round of 16
r16_matches_df = pd.DataFrame([
    {"match_id": "R16_1", "round": "Round of 16", "team1": "Senegal", "team2": "Sudan"},
    {"match_id": "R16_2", "round": "Round of 16", "team1": "Mali", "team2": "Tunisia"},
    {"match_id": "R16_3", "round": "Round of 16", "team1": "Morocco", "team2": "Tanzania"},
    {"match_id": "R16_4", "round": "Round of 16", "team1": "South Africa", "team2": "Cameroon"},
    {"match_id": "R16_5", "round": "Round of 16", "team1": "Egypt", "team2": "Benin"},
    {"match_id": "R16_6", "round": "Round of 16", "team1": "Nigeria", "team2": "Mozambique"},
    {"match_id": "R16_7", "round": "Round of 16", "team1": "Algeria", "team2": "DR Congo"},
    {"match_id": "R16_8", "round": "Round of 16", "team1": "Cote d'Ivoire", "team2": "Burkina Faso"},
])

# Quarterfinals
qf_matches_df = pd.DataFrame([
    {"match_id": "QF_1", "round": "Quarterfinal", "team1": "Mali", "team2": "Senegal"},
    {"match_id": "QF_2", "round": "Quarterfinal", "team1": "Egypt", "team2": "Cote d'Ivoire"},
    {"match_id": "QF_3", "round": "Quarterfinal", "team1": "Cameroon", "team2": "Morocco"},
    {"match_id": "QF_4", "round": "Quarterfinal", "team1": "Algeria", "team2": "Nigeria"},
])

# Semifinals
sf_matches_df = pd.DataFrame([
    {"match_id": "SF_1", "round": "Semifinal", "team1": "Senegal", "team2": "Egypt"},
    {"match_id": "SF_2", "round": "Semifinal", "team1": "Morocco", "team2": "Nigeria"},
])

# Final
final_matches_df = pd.DataFrame([
    {"match_id": "F_1", "round": "Final", "team1": "Senegal", "team2": "Morocco"},
])

# Combine all matches
all_matches_df = pd.concat(
    [group_matches_df, r16_matches_df, qf_matches_df, sf_matches_df, final_matches_df],
    ignore_index=True
)

# ============================================================
# 3) GREEDY HEURISTIC FUNCTION
# ============================================================

def run_greedy_assignment(matches_df, output_prefix="heuristic"):
    """
    Priority-based greedy assignment heuristic.

    Realism constraints:
    - Morocco always plays in Prince Moulay Abdellah Sports Complex (Rabat)
    - Semifinals and Final restricted to large venues only (capacity >= 45000)

    Outputs only:
    - {output_prefix}_assignments.csv
    - {output_prefix}_team_travel_times.csv
    """

    stadium_city = stadiums_df.set_index("stadium")["city"].to_dict()
    stadium_capacity = stadiums_df.set_index("stadium")["capacity"].to_dict()

    all_stadiums = stadiums_df["stadium"].tolist()
    large_stadiums = [s for s in all_stadiums if stadium_capacity[s] >= 45000]
    host_stadium = "Prince Moulay Abdellah Sports Complex"

    # Priority: later rounds first, Morocco matches first
    round_priority = {
        "Final": 1,
        "Semifinal": 2,
        "Quarterfinal": 3,
        "Round of 16": 4,
        "Group Stage": 5,
    }

    matches_df = matches_df.copy()
    matches_df["round_priority"] = matches_df["round"].map(round_priority).fillna(99)
    matches_df["morocco_priority"] = matches_df.apply(
        lambda r: 0 if "Morocco" in [r["team1"], r["team2"]] else 1,
        axis=1
    )
    matches_df = matches_df.sort_values(
        by=["round_priority", "morocco_priority", "match_id"]
    ).reset_index(drop=True)

    teams_in_set = sorted(set(matches_df["team1"]).union(set(matches_df["team2"])))
    team_total_travel = {t: 0.0 for t in teams_in_set}

    def feasible_stadiums_for_match(row_dict):
        if "Morocco" in [row_dict["team1"], row_dict["team2"]]:
            return [host_stadium]
        if row_dict["round"] in ["Semifinal", "Final"]:
            return large_stadiums
        return all_stadiums

    def round_trip_cost(team, stadium):
        base_city = team_base_city[team]
        venue_city = stadium_city[stadium]
        return 2 * travel_time[base_city][venue_city]

    assignment_rows = []
    team_match_rows = []

    for row_dict in matches_df.to_dict(orient="records"):
        feasible_stadiums = feasible_stadiums_for_match(row_dict)

        best_stadium = None
        best_candidate_max = float("inf")
        best_candidate_sum = float("inf")

        for s in feasible_stadiums:
            candidate_totals = team_total_travel.copy()
            candidate_totals[row_dict["team1"]] += round_trip_cost(row_dict["team1"], s)
            candidate_totals[row_dict["team2"]] += round_trip_cost(row_dict["team2"], s)

            candidate_max = max(candidate_totals.values())
            candidate_sum = sum(candidate_totals.values())

            if candidate_max < best_candidate_max:
                best_candidate_max = candidate_max
                best_candidate_sum = candidate_sum
                best_stadium = s
            elif candidate_max == best_candidate_max:
                if candidate_sum < best_candidate_sum:
                    best_candidate_sum = candidate_sum
                    best_stadium = s
                elif candidate_sum == best_candidate_sum:
                    if best_stadium is None or stadium_capacity[s] > stadium_capacity[best_stadium]:
                        best_stadium = s

        stadium = best_stadium
        city = stadium_city[stadium]

        t1_cost = round_trip_cost(row_dict["team1"], stadium)
        t2_cost = round_trip_cost(row_dict["team2"], stadium)

        team_total_travel[row_dict["team1"]] += t1_cost
        team_total_travel[row_dict["team2"]] += t2_cost

        assignment_rows.append({
            "Match ID": row_dict["match_id"],
            "Round": row_dict["round"],
            "Team 1": row_dict["team1"],
            "Team 2": row_dict["team2"],
            "Assigned Stadium": stadium,
            "Assigned City": city
        })

        team_match_rows.append({
            "Team": row_dict["team1"],
            "Base City": team_base_city[row_dict["team1"]],
            "Round": row_dict["round"],
            "Match ID": row_dict["match_id"],
            "Assigned Stadium": stadium,
            "Assigned City": city,
            "Travel Time for Match (hours)": t1_cost
        })

        team_match_rows.append({
            "Team": row_dict["team2"],
            "Base City": team_base_city[row_dict["team2"]],
            "Round": row_dict["round"],
            "Match ID": row_dict["match_id"],
            "Assigned Stadium": stadium,
            "Assigned City": city,
            "Travel Time for Match (hours)": t2_cost
        })

    assignments_df = pd.DataFrame(assignment_rows)
    team_travel_times_df = pd.DataFrame(team_match_rows)

    # Add stage total per team
    stage_totals = (
        team_travel_times_df
        .groupby(["Team", "Round"])["Travel Time for Match (hours)"]
        .sum()
        .rename("Stage Travel Total (hours)")
        .reset_index()
    )
    team_travel_times_df = team_travel_times_df.merge(
        stage_totals,
        on=["Team", "Round"],
        how="left"
    )

    # Add competition total per team
    competition_totals = (
        team_travel_times_df
        .groupby("Team")["Travel Time for Match (hours)"]
        .sum()
        .rename("Competition Travel Total (hours)")
        .reset_index()
    )
    team_travel_times_df = team_travel_times_df.merge(
        competition_totals,
        on="Team",
        how="left"
    )

    # Order nicely
    round_order = {
        "Group Stage": 1,
        "Round of 16": 2,
        "Quarterfinal": 3,
        "Semifinal": 4,
        "Final": 5
    }
    team_travel_times_df["round_order"] = team_travel_times_df["Round"].map(round_order).fillna(99)
    team_travel_times_df = team_travel_times_df.sort_values(
        by=["Team", "round_order", "Match ID"]
    ).drop(columns=["round_order"])

    # Save only two files
    assignments_file = f"{output_prefix}_assignments.csv"
    travel_file = f"{output_prefix}_team_travel_times.csv"

    assignments_df.to_csv(assignments_file, index=False)
    team_travel_times_df.to_csv(travel_file, index=False)

    # Print summary
    print("Current working directory:", os.getcwd())

    print(f"\n=== {output_prefix.upper()} ASSIGNMENTS ===")
    print(assignments_df.to_string(index=False))

    print(f"\n=== {output_prefix.upper()} TEAM TRAVEL TIMES ===")
    print(team_travel_times_df.to_string(index=False))

    max_total = team_travel_times_df.groupby("Team")["Competition Travel Total (hours)"].first().max()
    total_all = team_travel_times_df.groupby("Team")["Competition Travel Total (hours)"].first().sum()

    print(f"\nMax team competition travel: {max_total:.2f} hours")
    print(f"Total competition travel across teams: {total_all:.2f} hours")

    print("\nFiles created:")
    print(f"- {assignments_file}")
    print(f"- {travel_file}")

    return assignments_df, team_travel_times_df

# ============================================================
# 4) RUN THE HEURISTIC
# ============================================================

heuristic_assignments_df, heuristic_team_travel_times_df = run_greedy_assignment(
    all_matches_df,
    output_prefix="heuristic_all_competition"
)
```
